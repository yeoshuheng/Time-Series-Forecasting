Types of moving averages, SMA & EMA. EMA is faster to react to the markets, but more error prone, whereas SMA takes a longer time to react to the markets but is more immune to random, erratic behaviors in the markets. 125-Days: Overall market momentum, a rise in price over the 125-days moving average shows that there is good momentum to the stock.

50-Days: Long term moving average of the stock.
15-Days: Short term moving average of the stock.

# Data Processing 

```python
def bollinger_bands(df):
    """
    Takes in a dataframe with the SMA15 already calculated.
    Calculates and returns the same dataframe with the upper and lower bollinger bands.
    """
    
    # Calculate SMA20
    df['SMA20'] = df['Close'].rolling(20).mean()
    
    # Calculate standard deviation
    std = df['SMA20'].rolling(15).std()
    
    df['lowerbol'] = df['SMA20'] + std * 2
    df['upperbol'] = df['SMA20'] - std * 2
    
    return df
```

```python
def split(df, days_used_to_pred):
    """
    Takes in a dataframe and splits it into a training set and a testing set (1 set of timeframe for the model to predict.)
    It will then return the training set and testing set and the scaler function itself.
    """
    
    # Split data
    df_train = df
    df_test = df[-(days_used_to_pred + FINAL_DAYS_NEEDED):]
    
    return df_train, df_test
    
```

```python
def create_vars(df, timestep):
    """
    Takes in the data frame the time steps we are using to predict the values and the days to predict. 
    Normalises the independant variables before splitting and returning the independent and dependent variables as 2 seperate series.
    Returns the 2 seperate series together with the scaler used to normalised them.
    """   
    x = []
    y = []
    
    # Calculate the last set of days to be added to the list.
    end = len(df)- timestep
    
    # Standardise the dataframe
    scaler, df = standardise(df)
    
    # Split into x & y
    for i in range(end):
        
        # Gather the lists that are needed to extract x & y variables
        x_full = df[i : timestep + i]
        y_full = df[timestep + i]
        
        # Extract x and y variables from the lists gathered
        x_vars = [lst for lst in x_full]
            
        # Add the variables to the final x
        x.append(np.array(x_vars).astype(np.float32))
        
        # Add the variables to the final y
        y.append(np.array(y_full).astype(np.float32))
        
    # Convert to np arrays        
    np_x = np.array(x)
    np_y = np.array(y)
    
    # Convert to tensors 
    x = tf.convert_to_tensor(np_x, dtype = np.float32)
    y = tf.convert_to_tensor(np_y, dtype = np.float32)
    
    return scaler, x, y
```

```python
def standardise(df):
    """
    Takes in a data frame and standardises values.
    Returns the normalised data as a numpy data frame.
    """
    
    scaler = StandardScaler()
    df = scaler.fit_transform(df)
    
    return scaler, df
```

```python
# Create moving average
df['SMA125'] = df['Close'].rolling(125).mean()
df['SMA50'] = df['Close'].rolling(50).mean()

# Create the bollinger bands
df = bollinger_bands(df)

# Drop first 125 days (No MA values)
df = df[125:]
```

# Building the model

Choice of model:

Stacked LSTM

Deciding on the architecture of the LSTM model was done via trial and error. Initially I just used a normal stacked LSTM, however it became too over-generalised to the overall trend, (Returning flat predictions).

Hence Bidirectional layers were added.

Regarding the 'timestep' or the number of days used to train the model, I initially used 5 days. However, this led to incccurate results, I think this is because anything below 5 days causes the model to neglect the overall trend of the stock prices. For example, if the stock is on a overall down trend this month, the model would ignore that and predict a rise in stock prices over the next 10 days when it was just in fact a normal fluctuation.

I slowly increased the length of the time step and eventually settled on 15 days, which produced the best results, possibly because a balance of patterns from both the short-term and overall trend of the prices has been achieved.

```python
# Define some hyperparameters
EPOCHS = 100
INPUT_SHAPE = (DAYS_USED_TO_PRED, NO_VAR)
VALIDATION_SPLIT = 0.2
```

```python
# Define model
model = keras.Sequential([
    
    keras.layers.LSTM(64, return_sequences = True, input_shape = INPUT_SHAPE, activation = 'relu'),
    keras.layers.Bidirectional(
    keras.layers.LSTM(128, return_sequences = True, activation = 'relu')),
    keras.layers.LSTM(256, return_sequences = True, activation = 'relu'),
    keras.layers.Bidirectional(
    keras.layers.LSTM(128, return_sequences = True, activation = 'relu')),
    keras.layers.LSTM(64, input_shape = INPUT_SHAPE, activation = 'relu'),
    
    keras.layers.Dropout(0.2),
    keras.layers.Dense(30, activation = 'linear'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(20, activation = 'linear'),
    
    # Final Output
    keras.layers.Dense(NO_VAR, activation = 'linear')
    
])

model.compile(optimizer = 'adam',
             loss = 'mean_squared_error')

model.summary()
```

```python
# Train Model
x_train = np.array(x_train).reshape(-1, DAYS_USED_TO_PRED, NO_VAR)
model.fit(x_train, y_train, epochs = EPOCHS, validation_split = VALIDATION_SPLIT)
```

```python
def predict_y(n, df):
    """
    Takes in number of days for the model to predict and a numpy array used for prediction.
    Returns predictions for the next few days.
    """
    
    # Normalise the dataframe
    df = scaler_used.fit_transform(df)
    
    predictions = []
    
    # Iterate through the values
    for i in range(n):
        
        # Reshape the data for the prediction
        df = df.reshape(-1, DAYS_USED_TO_PRED, NO_VAR)
        
        # Generate prediction and add it to the prediction
        prediction = model.predict(df)
        predictions.append(prediction[0])
        
        # Create the next data frame used for the prediction
        
        # Drop first day from the data frame
        df = df[0][1:]
        
        # Add predicted next day to the data frame
        df = np.concatenate((df, prediction), axis = 0)

    return predictions
```

```python
def inverse(predictions):
    """
    Takes in a series of predictions and transforms them back to the original values.
    Returns the predictions as their original values.
    """
    
    # Convert rows of predictions into a dataframe
    df = pd.DataFrame(predictions)
    
    # Transforms the predictions back to original values
    df = scaler_used.inverse_transform(df)
    
    return df
```

```python
def gather_results(index, predictions):
    """
    Takes in index of results we want and the final transformed predictions.
    Returns the final transformed predictions in a list.
    """
    
    return [lst[index] for lst in predictions]
```
