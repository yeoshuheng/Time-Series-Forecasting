```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib as plot
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow import keras
```

# Data Collection 

We will use historical stock prices of NVIDIA for our predictions. NVIDIA is primarily involved in 5 markets - Gaming, Data Centers, Automotive Devices, OEM and visualisation.

Stock prices of individual companies are affected by the industry the company is in. Negative sentiment for the entire sector would lead drag the company down with it. 

Hence, apart from the overall market, we will also bring in data from NVIDIA's peers. This can be done by bringing in the Technology Select Sector SPDR Fund (XLK), which tracks the performance of the tech sector. Some of NVIDIA's biggest competitors, such as AMD and Intel are also included in the top 10 holdings of this ETF.

```python
# Global Variables
NO_VAR = 7
START_D = '2022-01-05'
END_D = '2022-07-10'
DAYS_TO_PRED = 1
DAYS_USED_TO_PRED = 3
```

```python
def get_data_yf(start, end, ticker = 'NVDA'):
    """
    get_data_yf takes in 3 strings, a ticker (Default being NVDA), a start date and end date.
    It returns the historical stock price data from yahoo finance over the period of the start and end date.
    """
    
    df = yf.download(
        ticker,
        start = start,
        end = end
    )
    
    return df
```

```python
# NVIDIA dataset
df = get_data_yf(START_D, END_D)
```

```python
# XLK dataset
df_xlk = get_data_yf(START_D, END_D, 'XLK')
```

```python
# VIX
df_vix = get_data_yf(START_D, END_D, '^VIX')
```

# Data Processing 

```python
# Check size
print('XLK:' + str(df_xlk.shape) + ', NVIDIA:' + str(df.shape) + ', VIX:' + str(df_vix.shape))

# Check null values
for ticker in [df_xlk, df, df_vix]:
    print(str(ticker.isnull().values.any()))
```

```python
# Combine XLK and NVIDIA into a singular dataframe.
df['XLK'] = df_xlk['Close']
df['NAS'] = df_vix['Close']
df.head()
```

We will not touch Adjusted Closing prices since that factors in stock splits and dividends, we will purely focus on the closing price.

```python
# Dropping the columns not needed.
df.drop(['Adj Close'], axis = 1, inplace = True)
df.head()
```

```python
def split(df, days_to_pred, days_used_to_pred):
    """
    Takes in a dataframe and splits it into a training set and a testing set (1 set of timeframe for the model to predict.)
    It will then return the training set and testing set and the scaler function itself.
    """
    
    # Split data
    df_train = df
    df_test = df[- (days_to_pred + days_used_to_pred):]
    
    return df_train, df_test
    
```

```python
def standardise(df):
    """
    Takes in a data frame and standardises values.
    Returns the normalised data as a numpy data frame.
    """
    
    scaler = StandardScaler()
    df = scaler.fit_transform(df)
    
    return scaler, df
```

```python
def create_vars(df, timestep, days_to_pred):
    """
    Takes in the data frame the time steps we are using to predict the values and the days to predict. 
    Normalises the independant variables before splitting and returning the independent and dependent variables as 2 seperate series.
    Returns the 2 seperate series together with the scaler used to normalised them.
    """   
    x = []
    y = []
    
    # Calculate the last set of days to be added to the list.
    end = len(df) - days_to_pred - timestep
    
    # Standardise the dataframe
    scaler, df = standardise(df)
    
    # Split into x & y
    for i in range(end):
        
        # Gather the lists that are needed to extract x & y variables
        x_full = df[i : timestep + i]
        y_full = df[timestep + i :timestep + i + days_to_pred]
        
        # Extract x and y variables from the lists gathered
        x_vars = [lst for lst in x_full]
        y_vars = [lst[0] for lst in y_full]
            
        # Add the variables to the final x
        x.append(np.array(x_vars).astype(np.float32))
        
        # Add the variables to the final y
        y.append(np.array(y_vars).astype(np.float32))
        
    # Convert to np arrays        
    np_x = np.array(x)
    np_y = np.array(y)
    
    # Convert to tensors 
    x = tf.convert_to_tensor(np_x, dtype = np.float32)
    y = tf.convert_to_tensor(np_y, dtype = np.float32)
    
    return scaler, x, y
```

```python
df_train, df_test = split(df, DAYS_TO_PRED, DAYS_USED_TO_PRED)
df_train
```

```python
scaler_used, x_train, y_train = create_vars(df_train, DAYS_USED_TO_PRED, DAYS_TO_PRED)

# Check that the split is conducted and x, y variable split successfully.
x_train[0]
```

```python
y_train[0]
```

```python
print("X variables: " + str(len(x_train)) + ", Y variables: " + str(len(y_train)))
print("X variables shape: " + str(x_train.shape) + ", Y variables shape: " + str(y_train.shape))
```

# Building the model

```python
# Define some hyperparameters
EPOCHS = 200
INPUT_SHAPE = (DAYS_USED_TO_PRED, NO_VAR)
```

```python
# Define model
model = keras.Sequential([
    keras.layers.LSTM(100,return_sequences = True, input_shape = INPUT_SHAPE, activation = 'relu'),
    keras.layers.LSTM(100, return_sequences = True, activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.LSTM(100, activation = 'relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(50),
    keras.layers.Dropout(0.3),
    
    # Final Output
    keras.layers.Dense(DAYS_TO_PRED)
    
])

model.compile(optimizer = 'adam',
             loss = 'mean_squared_error')

model.summary()
```

```python
# Train Model
x_train = np.array(x_train).reshape(-1, INPUT_SHAPE[0], INPUT_SHAPE[1])
model.fit(x_train, y_train, epochs = EPOCHS)
```

```python
# Take a look at movement of loss throughout training
loss = model.history.history['loss']
plt.plot(loss)
```

# Model Evaluation

```python
def inverse_scaler_y(y_pred, scaler):
    """
    Takes in array of scaled predicted y values together with a scaler function.
    Transforms the scaled y values into the original y values.
    Returns the original y values.
    """
    
    # Convert the predicted values into a data frame
    y_pred = pd.DataFrame(y_pred[0], columns = ['Close'])

    # Create copies of the same column so that inverse transform recognises the dimensions.
    for i in range(NO_VAR - 1):
        y_pred[str(i)] = y_pred['Close']
    
    # Inverse transform.
    y_pred = scaler.inverse_transform(y_pred)
    
    # Extract the original y-values.
    y_pred = [lst[0] for lst in y_pred]
    
    # Convert the original y-values into a dataframe again.
    y_pred = pd.DataFrame(y_pred, columns = ['Close'])
    y_pred['dates'] = y_test.index
    y_pred = y_pred.set_index('dates')
    
    return y_pred
```

```python
# Create test sets
x_test = df_test[:DAYS_USED_TO_PRED]
y_test = df_test[-DAYS_TO_PRED:]

# Normalise the x_test values with the same scaler used on the training set
x_test = scaler_used.fit_transform(x_test)

# Reshape x_test for the model
x_test = np.array(x_test).reshape(-1, DAYS_USED_TO_PRED, NO_VAR)
```

```python
# Check shape of the test sets
print("X_test shape: " + str(x_test.shape) + ", Y_test shape: " + str(y_test.shape))
```

```python
y_pred = model.predict(x_test)
```
